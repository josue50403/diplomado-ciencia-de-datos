## Paquete scikit learn (sklearn)
Cargamos los paquetes necesarios para el algoritmos, usamos datos por defecto en el paquete de datos. Iris data.


```python
#conda install -c conda scikit-learn
#conda install -c conda-forge 
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from sklearn.tree import DecisionTreeClassifier
from sklearn.datasets import load_iris #Hacemos uso de los datos iris
from sklearn import tree #Graficar el arbol
from sklearn.model_selection import train_test_split # Para separar grupo de test y traintin 
from sklearn.metrics import confusion_matrix, accuracy_score
from sklearn.neighbors import KNeighborsClassifier # El k vecions mas cercano
```

```python
iris = load_iris()
#Hacemos selecion, grupo de test y training
X_train, X_test, Y_train, Y_test = train_test_split(iris.data, iris.target, test_size=0.25,  random_state=0)
```

```python
cfl = DecisionTreeClassifier(max_leaf_nodes=5, random_state=0)
```

### Ajustamos el modelo Clasificador de Arbol de decision

```python
cfl.fit(X_train, Y_train)
```

```python
tree.plot_tree(cfl)
```

Ahora vamos ver indicadores de rendimiento.
Este dataset es un ideal, se realiza con el 25% de grupo de test

```python
accuracy=cfl.score(X_test, Y_test)
accuracy
```

```python
Y_pred=cfl.predict(X_test)
matriz_confusion=confusion_matrix(Y_test, Y_pred)
matriz_confusion
```
